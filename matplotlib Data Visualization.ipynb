```python
#Module 6: Assignment 1 - Data Visualization
#2. Generate a line plot graph for the column ‘model’ and ‘hp’.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


cars= pd.read_csv(r"D:\@NIL\downloads\intellipaat\assignments\2adv python\MODULE-DATA-VISUALIZATION-USING-MATPLOTLIB-ASSIGNMENT-02SEP2022151321\Matplotlib-2-1\cars-1.csv")
cars
```

```python
#fig=plt.figure(figsize=(10,2))
cars.plot('model','hp')
plt.grid()
plt.title("Model Names vs HorsePower", fontname='Comic Sans MS', fontsize=18)
plt.xlabel("Models of the cars", fontname='Arial', fontsize=12)
plt.ylabel("Horse-Power of Cars", fontname='sans-serif', fontsize=12)
plt.show()
```

```python
#Module 6: Assignment 2 - Data Visualization


#1. Generate a bar plot graph for the columns ‘carbs’ and ‘gear’.
fig=plt.figure(figsize=(10,5))
#ax1=plt.subplot()
plt.bar(np.array(cars['carb']),np.array(cars['gear']))
plt.grid()
plt.title("carbs vs gear", fontname='Comic Sans MS', fontsize=18)
plt.xlabel("Number of carburetors", fontname='Arial', fontsize=12)
plt.ylabel("Number of forward gears", fontname='sans-serif', fontsize=12)
plt.show()
```

```python
#Module 6: Assignment 3 - Data Visualization

#1. Plot a histogram for the column ‘wt’.
fig=plt.figure(figsize=(10,5))
plt.hist(np.array(cars['wt']),bins=30)
plt.grid()
plt.xticks(ticks=range(0,5,1))
plt.title("Histogram for the weight values", fontname='Comic Sans MS', fontsize=18)
plt.xlabel("weight of the cars", fontname='Arial', fontsize=12)
plt.ylabel("Count", fontname='sans-serif', fontsize=12)
plt.show()
```

```python
#Module 6: Assignment 4 - Data Visualization

#1. Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data frame.
#cars= pd.read_csv(r"D:\@NIL\downloads\intellipaat\assignments\2adv python\MODULE-DATA-VISUALIZATION-USING-MATPLOTLIB-ASSIGNMENT-02SEP2022151321\Matplotlib-2-1\cars-1.csv")


plt.pie(x=(np.array(cars['cyl'])),labels=(np.array(cars['model'])))
plt.show()
```

```python
#Module 6: Assignment 5 - Data Visualization

#1. Plot the area chart for the colmns: ‘am’ and ‘carb’.

fig=plt.figure(figsize=(10,5))
plt.stackplot(np.array(cars['am']),np.array(cars['carb']))
plt.grid()
plt.xticks(ticks=range(0,1))
plt.title(" Transmission vs Number of carburetors.", fontname='Comic Sans MS', fontsize=18)
plt.xlabel("Transmission", fontname='Arial', fontsize=12)
plt.ylabel("Number of carburetors", fontname='sans-serif', fontsize=12)
plt.show()
```
